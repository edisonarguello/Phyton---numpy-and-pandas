# merge

```python
izq=pd.DataFrame({'Key':['K1','K2','K3'],'A':['A0','A1','A3'],'B':['B0','B1','B3'],'C':['C0','C1','C3']})
izq
```

```python
der=pd.DataFrame({'Key':['K1','K2','K3'],'D':['D0','D1','D3'],'E':['E0','E1','E3'],'F':['F0','F1','F3']})
```

```python
izq.merge(der)
```

```python
der2=pd.DataFrame({'Key2':['K1','K2','K3'],'D':['D0','D1','D3'],'E':['E0','E1','E3'],'F':['F0','F1','F3']})
der2
```

```python
izq.merge(der2,left_on='Key',right_on='Key2')
```

```python
der3=pd.DataFrame({'Key2':['K1','K2','K3'],'D':['D0','D1','D3'],'E':['E0','E1','E3'],'F':['F0','F1',np.nan]})
```

```python
izq.merge(der3,left_on='Key',right_on='Key2')
```

# join

```python
izq10=pd.DataFrame({'A':['A0','A1','A3'],'B':['B0','B1','B3'],'C':['C0','C1','C3']},index=['K0','K1','K2'])
```

```python
izq10
```

```python
der10=pd.DataFrame({'D':['D0','D1','D3'],'E':['E0','E1','E3'],'F':['F0','F1','F3']},index=['K0','K2','K3'])
der10
```

```python
izq10.join(der10,how='inner')
```

```python
izq10.join(der10,how='right')
```

```python
izq10.join(der10,how='left')
```

```python
izq10.join(der10,how='outer')
```
